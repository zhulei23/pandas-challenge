```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players

```python
purchase_data_df = pd.DataFrame(purchase_data)
purchase_data_df.head()
```

```python
# Collecting data for Total Players
df = purchase_data
df.head()
total_players = df['SN'].nunique()

# Create dataframe for Total Players result
total_players_summary_df = pd.DataFrame({"Total Players": [total_players]})
total_players_summary_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Collecting data for Number of Unique Items
number_of_unique_items = df['Item Name'].nunique()

# Collecting data for Average Price 
average_price = round(df['Price'].mean(),2)

# Collecting data for Number of Purchases
number_of_purchases = df['Purchase ID'].count()

# Collecting data for Total Revenue
total_revenue = df['Price'].sum()

# Create a summary DataFrame using above value
purchase_analysis_summary_df = pd.DataFrame({"Number of Unique Items": [number_of_unique_items],
                                            "Average Price": f'${average_price}',
                                            "Number of Purchases": [number_of_purchases],
                                            "Total Revenue": f'${total_revenue}'})
purchase_analysis_summary_df
```

## Gender Demographics

* Percentage and Count of Male Players
* Percentage and Count of Female Players
* Percentage and Count of Other / Non-Disclosed

```python
# Grouping data of Gender by calculating SN column
grouped_gender_df = df.groupby('Gender')['SN'].nunique()
grouped_gender_df.head()

# Calculate data for Percentage of Players
percentage_of_players = round((grouped_gender_df/total_players)*100,2)
percentage_of_players

# Merging two dataframes by using Inner join
merge_df = pd.merge(grouped_gender_df, percentage_of_players, how='inner', on='Gender')
merge_df

# Rename columns to 'Total Count' & 'Percentage of Players'
gender_demographics_df = merge_df.rename(columns={"SN_x": "Total Count", "SN_y":"Percentage of Players"})
gender_demographics_df.head()

# Rename index
gender_demographics_df.index = ["Female", "Male", "Other / Non-Disclosed"]
gender_demographics_df

# Adding % sign onto Percentage of Players column
gender_demographics_df['Percentage of Players'] = gender_demographics_df['Percentage of Players'].astype(str) + '%'
gender_demographics_df

# Sorting data in ascending order
gender_demographics_df.sort_values(by='Total Count', ascending=False)
```

## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Calculate data for Purchase Count
purchase_count = df.groupby('Gender')['SN'].count()
purchase_count

# Calculate Total Purchase Value
total_purchase_value = df.groupby('Gender')['Price'].sum()
total_purchase_value

# Calculate Average Purchase Price
average_purchase_price = round(total_purchase_value/purchase_count, 2)
average_purchase_price

# Calculate Avg Total Purchase Per Person
avg_total_purchase_per_person = round(total_purchase_value/grouped_gender_df,2)
avg_total_purchase_per_person

# Create dataframes which conclude all the calculations above & add $ sign on the final results
pd.options.display.float_format = '${:,.2f}'.format
purchase_analysis_gender = pd.DataFrame({"Purchase Count": purchase_count,
                                            "Average Purchase Price": average_purchase_price,
                                            "Total Purchase Value": total_purchase_value,
                                            "Avg Total Purchase per Person": avg_total_purchase_per_person})
purchase_analysis_gender
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# define dataframe for age_bin
age_df = purchase_data

# Create bins to hold values for ages
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 9999]

# Create the labels for the bins
group_labels = ["<10", "10-14", "15-19","20-24","25-29","30-34","35-39", "40+"]

# Slice the data and place it into bins
pd.cut(age_df["Age"], bins, labels=group_labels).head()

# Categorize the existing players using the age bins
age_df["Age Group"] = pd.cut(age_df["Age"], bins, labels=group_labels)
age_df.head()

# Calculate Total Count for each age group
age_group = age_df.groupby("Age Group")
age_group_count = age_group["SN"].nunique()

# Calculate Total count for all unique players
total_unique_players = age_group_count.sum()
total_unique_players

# Calculate Percentage of Players for each age group
percentage_of_players = round((age_group_count/total_unique_players) * 100,2)
percentage_of_players

# Create a summary data frame to hold the calculated results
age_demographics_summary = pd.DataFrame({"Total Count": age_group_count,
                                        "Percentage of Players": percentage_of_players})
age_demographics_summary

# Remove header on index
age_demographics_summary.index = ["<10", "10-14", "15-19","20-24","25-29","30-34","35-39", "40+"]
age_demographics_summary

# Adding % sign onto Percentage of Players column
age_demographics_summary['Percentage of Players'] = age_demographics_summary['Percentage of Players'].astype(str) + '%'
age_demographics_summary
```
